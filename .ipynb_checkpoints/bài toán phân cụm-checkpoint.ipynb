## Mô tả dữ liệu 

**Các cột trong dữ liệu:**
- age (int): Tuổi của khách hàng.
- job (object): Nghề nghiệp của khách hàng (ví dụ: admin., technician, services,...).
- marital (object): Tình trạng hôn nhân của khách hàng (single, married, divorced).
- education (object): Trình độ học vấn (primary, secondary, tertiary, unknown).
- default (object): Khách hàng có khoản nợ mặc định không? (yes, no).
- balance (int): Số dư trung bình hàng năm (EUR).
- housing (object): Khách hàng có khoản vay thế chấp không? (yes, no).
- loan (object): Khách hàng có khoản vay cá nhân không? (yes, no).
- contact (object): Loại thông tin liên lạc (cellular, telephone, unknown).
- day (int): Ngày trong tháng mà khách hàng được liên hệ.
- month (object): Tháng khách hàng được liên hệ (ví dụ: jan, feb, mar,...).
- duration (int): Thời lượng của cuộc gọi (giây).
- campaign (int): Số lần liên hệ thực hiện trong chiến dịch hiện tại với khách hàng.
- pdays (int): Số ngày kể từ lần liên hệ cuối cùng trong chiến dịch trước đó (giá trị -1 nếu khách hàng chưa được liên hệ).
- previous (int): Số lần liên hệ với khách hàng trong chiến dịch trước đó.
- poutcome (object): Kết quả của chiến dịch tiếp thị trước đó (success, failure, other, unknown).
- deposit (object): Khách hàng có mở tài khoản tiền gửi không? (yes, no).

**Đặc điểm dữ liệu:**
- Không có giá trị bị thiếu: Số lượng giá trị bị thiếu trong tất cả các cột là 0.


## tiền sử lý dữ liệu

```python
#kiểm tra giữ liệu trùng lặp
data.duplicated().sum()
```

```python
# Phát hiện ngoại lai bằng phương pháp IQR
def detect_outliers_iqr(data, feature):
    Q1 = data[feature].quantile(0.25)
    Q3 = data[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[feature] < lower_bound) | (data[feature] > upper_bound)]
    return outliers

# Kiểm tra ngoại lai trong các cột
outliers_age = detect_outliers_iqr(data, 'age')
outliers_balance = detect_outliers_iqr(data, 'balance')

# In số lượng ngoại lai
print(f"Outliers in 'age': {outliers_age.shape[0]}")
print(f"Outliers in 'balance': {outliers_balance.shape[0]}")

# Loại bỏ ngoại lai
data_cleaned = data
```

```python
data_cleaned.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

# Khởi tạo LabelEncoder
label_encoder = LabelEncoder()

# Áp dụng LabelEncoder cho các cột phân loại
data_cleaned['job'] = label_encoder.fit_transform(data_cleaned['job'])
data_cleaned['marital'] = label_encoder.fit_transform(data_cleaned['marital'])
data_cleaned['education'] = label_encoder.fit_transform(data_cleaned['education'])
data_cleaned['deposit'] = label_encoder.fit_transform(data_cleaned['deposit'])
data_cleaned['housing'] = label_encoder.fit_transform(data_cleaned['housing'])
data_cleaned['loan'] = label_encoder.fit_transform(data_cleaned['loan'])
data_cleaned['contact'] = label_encoder.fit_transform(data_cleaned['contact'])
data_cleaned['poutcome'] = label_encoder.fit_transform(data_cleaned['poutcome'])
data_cleaned['default'] = label_encoder.fit_transform(data_cleaned['default'])
```

```python
data_cleaned
```

```python
from sklearn.preprocessing import MinMaxScaler
columns_to_scale = ['age', 'balance', 'duration']

# Lấy các cột cần chuẩn hóa từ dữ liệu
X_to_scale = data_cleaned[columns_to_scale].values

# Áp dụng MinMaxScaler chỉ cho các cột này
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_to_scale)
```

```python
scaler.fit(X_to_scale)
```

```python
other_columns = data_cleaned.drop(columns=columns_to_scale)  # Loại bỏ các cột đã chuẩn hóa
dt = pd.concat([pd.DataFrame(X_scaled, columns=columns_to_scale), other_columns.reset_index(drop=True)], axis=1)
```

```python
dt
```

```python
# Xóa một cột duy nhất
df = dt.drop('month', axis=1)

# Kiểm tra lại DataFrame sau khi xóa cột
print(df.head())
```

```python
# Tính ma trận tương quan
corr_matrix = df.corr()

# Hiển thị ma trận tương quan
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.show()

```

```python
from sklearn.ensemble import RandomForestClassifier
features = df.drop('deposit', axis=1).columns
# Dựng mô hình RandomForest để đánh giá độ quan trọng của các tính năng
model = RandomForestClassifier()
model.fit(df[features], df['deposit'])  # 'target' là cột bạn dự đoán

# Hiển thị độ quan trọng của các tính năng
importances = model.feature_importances_
feature_importance = pd.Series(importances, index=features)
feature_importance = feature_importance.sort_values(ascending=False)

# In ra độ quan trọng của từng tính năng
print(feature_importance)
```

```python
# Danh sách các đặc trưng cần giữ lại (chọn những đặc trưng có độ quan trọng cao)
important_features = [
    'duration', 'balance', 'age', 'day', 'pdays', 'job', 'campaign', 
    'contact', 'poutcome', 'housing', 'education', 'previous', 'deposit'
]

# Giả sử 'data_cleaned' là dữ liệu đã được chuẩn hóa và có các cột bạn muốn giữ lại
# Loại bỏ các cột không quan trọng
data_cleaned_reduced = df[important_features]

# Kiểm tra lại dữ liệu đã giảm bớt cột
print(data_cleaned_reduced.head())
```

```python
featuress = data_cleaned_reduced.drop('deposit', axis=1).columns
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Danh sách các giá trị K mà bạn muốn thử nghiệm
k_range = range(1, 11)

# Danh sách để lưu giá trị WCSS (hoặc SSE) cho từng giá trị K
wcss = []

# Áp dụng K-means cho từng giá trị K
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_cleaned_reduced[featuress])  # data_cleaned[features] là dữ liệu bạn sẽ phân cụm
    wcss.append(kmeans.inertia_)  # inertia_ là tổng sai số trong cụm (WCSS)

# Vẽ đồ thị Elbow
plt.figure(figsize=(8, 6))
plt.plot(k_range, wcss, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.show()
```

K = 3 là số cụm tối ưu mà bạn nên chọn, vì sau điểm này, sự giảm WCSS bắt đầu chậm lại và không có sự thay đổi rõ ràng.
Việc chọn K = 3 giúp bạn có 3 nhóm khách hàng khác nhau, với các đặc điểm phân biệt nhưng không quá phức tạp.

```python
#Gọi thư viện thuật toán phân cụm KMeans:
from sklearn.cluster import KMeans

#Thực hiện phân thành 3 cụm, các tham số khác sử dụng mặc định:
model_km = KMeans(n_clusters=4)

#Liệt kê các tham số của mô hình:
model_km.get_params()
```

```python
data_cleaned_reduced.info()
```

```python
df_new = data_cleaned_reduced.drop(columns=['deposit'])
```

```python
X = df_new[['balance', 'age', 'duration', 'campaign', 'day', 'pdays', 'job', 'contact', 'poutcome', 'housing', 'education', 'previous']].values
```

```python
model_km.fit_predict(X)

#Tạo cột Cluster để cho biết mẫu thuộc cụm nào
df_new["Cluster"] = model_km.labels_
df_new
```

```python
#Vị trí tâm của các cụm:
centorid = model_km.cluster_centers_
print('Tọa độ tâm các cụm:')
print(centorid)
```

```python
#Thống kê số lượng dữ liệu theo từng cụm:
sum_cluster = df_new['Cluster'].value_counts()
label =['Cụm 0','Cụm 1', 'Cụm 2','Cụm 3']

#Trực quan hóa dữ liệu
plt.figure(figsize=(10,5))
plt.bar(label, sum_cluster.values)

#Hiển thị nhãn dữ liệu trên mỗi cột
for i, v in enumerate(sum_cluster.values):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')
    
plt.title('THỐNG KÊ SỐ LƯỢNG THEO TỪNG CỤM')
plt.grid(ls='--')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Giả sử dữ liệu của bạn là X và labels là nhãn phân cụm từ KMeans

labels = df_new['Cluster']  # Nhãn phân cụm

# Giảm chiều dữ liệu xuống 2 chiều bằng PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X)

# Tạo DataFrame cho các thành phần PCA
pca_df = pd.DataFrame(data=pca_components, columns=['PCA1', 'PCA2'])
pca_df['Cluster'] = labels

# Vẽ biểu đồ phân cụm
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c=pca_df['Cluster'], cmap='viridis', s=50)
plt.title('KMeans Clustering with PCA')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
# Tính Silhouette Score
silhouette_avg = silhouette_score( data_cleaned_reduced, model_km.labels_)

# Tính Davies-Bouldin Index
davies_bouldin = davies_bouldin_score( data_cleaned_reduced, model_km.labels_)

# In kết quả
print(f"Silhouette Score: {silhouette_avg}")
print(f"Davies-Bouldin Index: {davies_bouldin}")
```

Kết quả bạn nhận được cho Silhouette Score và Davies-Bouldin Index cho thấy phân cụm của bạn khá tốt:

Silhouette Score = 0.81:
Giá trị này gần 1, cho thấy các cụm được phân tách khá rõ ràng và các điểm trong mỗi cụm có sự tương đồng cao.
Davies-Bouldin Index = 0.44:
Giá trị này thấp cho thấy các cụm không chỉ nhỏ gọn mà còn có sự tách biệt tốt. Thông thường, giá trị càng thấp càng tốt.
Nhận xét:
Cả hai chỉ số đều cho thấy phân cụm của bạn có chất lượng tốt.
Với Silhouette Score cao và Davies-Bouldin Index thấp, bạn có thể tự tin rằng mô hình phân cụm K-Means hiện tại đã đạt được sự tách biệt rõ ràng giữa các cụm.
